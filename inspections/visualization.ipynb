## `ydata_profiling` on each year

```python
from genericpath import exists


cols = [
    'StationId',
    'Type',
    'VIN',
    'Date',
    'MotorType',
    'Make',
    'VehicleType',
    'Model',
    'VehicleClass',
    'FirstRegistrationDate',
    'Km',
    # 'Defects',
    'Result',
    "Weekday",
    "DefectsA",
    "DefectsB",
    "DefectsC",
    "Defects0",
    "Defects1",
    "Defects2",
    "Defects3",
    "Defects4",
    "Defects5",
    "Defects6",
    "Defects7",
    "Defects8",
    "Defects9",
    "AgeDays"
]

dtypes = {
    'StationId': 'object',
    # 'Type': 'category',
    # 'Make': 'category',
    # 'Model': 'category',
    # 'MotorType': 'category',
    # 'VehicleClass': 'category',
    # 'Result': 'category',
    # "Weekday": 'int8',
    # "DefectsA": 'int8',
    # "DefectsB": 'int8',
    # "DefectsC": 'int8',
    # "Defects0": 'int8',
    # "Defects1": 'int8',
    # "Defects2": 'int8',
    # "Defects3": 'int8',
    # "Defects4": 'int8',
    # "Defects5": 'int8',
    # "Defects6": 'int8',
    # "Defects7": 'int8',
    # "Defects8": 'int8',
    # "Defects9": 'int8',
}

def load_year(year):
    df_all = None
    
    for name in [f'{year}-{month:02}' for month in range(1, 13)]:
        print(f'Load: {name}')
        if exists(f'{data_path}nosync/data_csv/{name}.csv'):
            df = pd.read_csv(f'{data_path}nosync/data_csv/{name}.csv', dtype=dtypes, usecols=cols, index_col=0)
            df['Date'] = pd.to_datetime(df['Date'])
            df['FirstRegistrationDate'] = pd.to_datetime(df['FirstRegistrationDate'])
        
        if df_all is None:
            df_all = df
        else:
            df_all = pd.concat([df_all, df])

    return df_all
```

```python
for year in [
        # '2018',
        '2019',
        '2020',
        '2021',
        '2022'
    ]:
    print(f'-------- {year} --------')
    df = load_year(year)
    profile = ProfileReport(df, title=f'Prohlídky na STK - {year}')
    profile.to_file(f'{data_path}exploratory_analysis/cleaned/inspections_report_{year}.html')
```

## How often does a single vehicle appear?

```python
# Load only the VIN and DatKont from all CSVs
from genericpath import exists


all_vins = None
for year in ['2018', '2019', '2020', '2021', '2022']:
    for name in [f'{year}-{month:02}' for month in range(1, 13)]:
        if exists(f'{data_path}data_csv/{name}.csv'):
            print(f'  {name}')
            vins = pd.read_csv(f'{data_path}data_csv/{name}.csv', usecols=['VIN'])
            if all_vins is None:
                all_vins = vins
            else:
                all_vins = pd.concat([all_vins, vins])
```

```python
display(all_vins)
```

```python
# Analyze uniqueness
vc = all_vins.value_counts()
vcc = vc.value_counts()
```

```python
vcc_simple = vcc.copy()
vcc_simple = vcc_simple.head(30)
fig, ax = plt.subplots(figsize=(10, 8))
g = sns.barplot(x=vcc_simple, y=vcc_simple.index, orient='h')
g.set_xscale("log")
g.set(title="Frequency of VIN repetition", xlabel="observations", ylabel="repetitions of a single VIN")
ax.bar_label(ax.containers[0])
g.plot()
```

## How polluted are some values?
Examine `ObchOznTyp` to determine how much information we will lose if we clean up the column. Focus on `OCTAVIA` as it's one of the most common values.

```python
def unique_freqs(dataset, col, h=10):
    df = pd.read_csv(f'{data_path}data_csv/{dataset}.csv', usecols=['ObchOznTyp'])
    octas = df[df['ObchOznTyp'].str.contains(col) == True]
    octas = octas['ObchOznTyp'].to_numpy()
    vals, counts = np.unique(octas, return_counts=True)

    fig, ax = plt.subplots(figsize=(10, h))
    g = sns.barplot(x=counts, y=vals, orient='h')
    # g.set_xscale("log")
    g.set(title=f"Frequency of different {col} subtypes", xlabel="observations", ylabel="subtypes")
    ax.bar_label(ax.containers[0])
    g.plot()
```

```python
unique_freqs('2021-06', 'OCTAVIA')
unique_freqs('2021-06', 'GOLF', h=15)
```

## How long are VINs?

```python
vins_lenghts = all_vins['VIN'].apply(lambda x: len(x))
vals, counts = np.unique(vins_lenghts, return_counts=True)

fig, ax = plt.subplots(figsize=(10, 10))
g = sns.barplot(x=counts, y=vals, orient='h')
# g.set_xscale("log")
g.set(title=f"Frequency of VIN lengths",
      xlabel="observations", ylabel="lenght")
ax.bar_label(ax.containers[0])
ax.set_xscale('log')
g.plot()
```
